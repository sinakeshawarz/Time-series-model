```python
import pandas as pd

url = 'https://raw.githubusercontent.com/sinakeshawarz/Time-series-model/main/ROKU.csv'
data = pd.read_csv(url)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# Let's assume that we are using only one feature, the 'High' price of the stock, to predict the 'Close' price. 
# You may choose multiple features based on your requirements.

X = data['High'].values.reshape(-1,1) 
y = data['Close'].values.reshape(-1,1)
```

```python
#Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#We'll train the Linear Regression, Decision Tree, and XGBoost regressors on your data:


# Linear Regression
regressor = LinearRegression()  
regressor.fit(X_train, y_train) # Training the algorithm

# Decision Tree
tree_regressor = DecisionTreeRegressor()  
tree_regressor.fit(X_train, y_train)

# XGBoost
xgb_regressor = XGBRegressor()
xgb_regressor.fit(X_train, y_train)
```

```python
#Long Short-Term Memory (LSTM) networks are a type of recurrent neural network (RNN) that can learn and remember over long sequences of information. They are especially good at sequence prediction problems because they can store past information. This is important in our case because the previous price of a stock is crucial in predicting its future price.

#We will be using the Keras library for creating the LSTM model. 

#We need to import some additional libraries for this part:

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
```

```python
#For LSTM, it is a good practice to scale the data to the range of 0-1. We will use MinMaxScaler from sklearn.preprocessing.
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1,1))

# We'll use 60 days stock price to predict the next day price
prediction_days = 60

x_train = []
y_train = []

for x in range(prediction_days, len(scaled_data)):
    x_train.append(scaled_data[x-prediction_days:x, 0])
    y_train.append(scaled_data[x, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
#Build the LSTM Model
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=25))
model.add(Dense(units=1))
```

```python
#Train the LSTM Model
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=25, batch_size=32)
```

```python
#Test the LSTM Model


# Let's assume you're using the last 20% of the data as test data
test_data_length = int(len(scaled_data) * 0.2)

# Create test data
test_data = scaled_data[len(scaled_data) - test_data_length - prediction_days:]
x_test = []

for x in range(prediction_days, len(test_data)):
    x_test.append(test_data[x-prediction_days:x, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))


predicted_prices = model.predict(x_test)
predicted_prices = scaler.inverse_transform(predicted_prices)
```

```python
#Visualize the Prediction
plt.figure(figsize=(10,5))
plt.plot(data['Close'].values, color='blue', label='Actual ROKU Stock Price')
plt.plot(range(len(y_train), len(y_train) + len(predicted_prices)), predicted_prices, color='red', label='Predicted ROKU Stock Price')
plt.title('ROKU Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('ROKU Stock Price')
plt.legend()
plt.show()
```

```python
#You can predict on the test set and evaluate the models using metrics like Mean Absolute Error, Mean Squared Error, and R2 score
# For Linear Regression
y_pred_lin_reg = regressor.predict(X_test)

print('Linear Regression')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_lin_reg))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_lin_reg))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_lin_reg)))

# For Decision Tree
y_pred_tree = tree_regressor.predict(X_test)

print('Decision Tree')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_tree))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_tree))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_tree)))

# For XGBoost
y_pred_xgb = xgb_regressor.predict(X_test)

print('XGBoost')
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_xgb))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred_xgb))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_xgb)))

```

```python
#Each of these plots will give you a visual representation of how well each model's predictions match the actual values. Points along the diagonal line indicate perfect predictions, while points that are off the diagonal represent errors.
# For Linear Regression
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred_lin_reg)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Linear Regression: Actual vs Predicted')
plt.grid(True)
plt.show()

# For Decision Tree
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred_tree)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Decision Tree: Actual vs Predicted')
plt.grid(True)
plt.show()

# For XGBoost
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred_xgb)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('XGBoost: Actual vs Predicted')
plt.grid(True)
plt.show()
```
